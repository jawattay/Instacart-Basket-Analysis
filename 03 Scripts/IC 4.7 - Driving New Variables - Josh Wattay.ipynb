### 3. Define price_label Function and Variable

```python
# If the value within the “prices” column within the given row is less than or equal to 5, then return the string “Low-range product.”

# Or else, if the value within the “prices” column within the given row is greater than 5 and less than or equal to 15, then return the string “Mid-range product.”

# Or else, if the value of the “prices” column within the given row is greater than 15, then return the string “High-range product.”

# Or else, return the string “Not enough data.”
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

### 4. Apply price_label Function to Subset

```python
# Applying new price_label function to subset of data

df['price_range'] = df.apply(price_label, axis=1)
```

#### Step 4
##### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

##### Lets start by getting value counts of the orders_time_of_day column to identify busiest and slowest hours first.

```python
#Frequency check of orders_time_of_day

ords_prods_merge['orders_time_of_day'].value_counts(dropna = False)
```

##### Given that the stakeholders want 3 groups and we have 24 hours, it makes sense that each group should have 8 individual hours assigned.
##### "Most Orders" will be: 10, 11, 14, 15, 13, 12, 16, 9
##### "Average Orders" will be: 17, 8, 18, 19, 20, 7, 21, 22
##### "Fewest Orders" will be: 6, 0, 1, 5, 2, 4, 3

```python
#Create shell Result prior to For-Loop code
result = []
```

```python
for value in ords_prods_merge["orders_time_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
    result.append("Most Orders")
  elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
    result.append("Average Orders")
  else:
    result.append("Fewest Orders")
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```
